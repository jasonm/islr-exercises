```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./Data/Default.csv")
df['student'] = df['student'].map(lambda x: 1 if x == 'Yes' else 0)
df['default'] = df['default'].map(lambda x: 1 if x == 'Yes' else 0)
```

```python
df.head()
```

```python
# DataFrame.plot.scatter(x, y, s=None, c=None, **kwds)


df.plot(kind='scatter', x='balance', y='income', c='default')
```

```python
from sklearn import linear_model
clf_balance = linear_model.LogisticRegression(C=1e5)
clf_balance.fit(df.balance.reshape(-1,1), df.default)

print "Balance"
print(clf_balance)
print('classes: ',clf_balance.classes_)
print('coefficients: ', clf_balance.coef_)
print('intercept :', clf_balance.intercept_)
```

```python
from sklearn import linear_model
clf = linear_model.LogisticRegression(C=1e5)
clf.fit(df.student.reshape(-1,1), df.default)

print(clf)
print('classes: ',clf.classes_)
print('coefficients: ',clf.coef_)
print('intercept :', clf.intercept_)
```
